```python
import threading
```

```python
def thread_fn(name):
    for i in range(10):
        time.sleep(1)
        print("{} running : {}".format(name, i))
        
def thread_fn2(name):
    i = 0
    while(True):
        time.sleep(1)
        print("{} running : {}".format(name, i))
        i+=1
        
def start_thread(name):
    thread = threading.Thread(target=thread_fn, args=(name, ))
    thread.start()
    return thread
    
def start_daemon(name):
    thread = threading.Thread(target=thread_fn2, args=(name, ), daemon=True)
    thread.start()
```
